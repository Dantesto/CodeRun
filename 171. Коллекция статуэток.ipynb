# Условие

Дан список чисел от $1$ до $n$. Необходимо найти в нем непрерывную последовательность, содержащую числа от $1$ до $k$, причем сумма этой последовательности должна быть минимальна.

# Решение

Это задача на 2 указателя.
<br>
Будем двигать правый указатель $r$ до тех пор, пока в диапазоне $[l, r]$ не окажутся все числа от $1$ до $k$, по пути суммируя все встречаемые числа. Потом начнем двигать левый указатель $l$ до тех пор, пока из диапазона не уйдет одно число от $1$ до $k$, по пути вычитая из суммы встречаемые числа.
<br>
Всякий раз, пока в диапазоне есть все нужные числа, обновляем минимальную сумму, сравнивая её с текущей.
<br>
Как отслеживать наличие нужных чисел в диапазоне? Для этого заведем список $cnt\_of\_num$ с индексами от $1$ до $k$, в котором будем увеличивать на $1$ значение $cnt\_of\_num[i]$ при встрече числа $i$ во время увеличения диапазона (движение правого указателя). Всякий раз, когда $cnt\_of\_num[i]$ из значения $0$ становится $1$, мы понимаем, что в диапазоне появилось ранее отсутсвовавшее число. Тогда счетчик $full$ уникальных чисел внутри диапазона будем увеличивать на $1$. Мы поймем, что в диапазоне есть все нужные числа, когда $full=k$. В этот момент мы и начнем двигать левую границу.
<br>
При ее движении обновление $cnt\_of\_num[i]$ и $full$ будет происходить аналогично, только уменьшаться.

# Код

```python
n, k = [int(x) for x in input().split()]
a = [int(x) for x in input().split()]
l, r = 0, 0
cnt_of_num = [0] * (k + 1) #cnt_of_num[i] = кол-во чисел i в текущем диапазоне [l,  r]
full = 0 #количество уникальных чисел от 1 до k в текущем диапазоне [l, r]
sum, min_sum = 0, float('inf')
for r in range(n):
    if a[r] <= k:
        if cnt_of_num[a[r]] == 0:
            full += 1
        cnt_of_num[a[r]] += 1
    sum += a[r]
    if full == k:
        while full == k:
            min_sum = min(min_sum, sum)
            if a[l] <= k:
                if cnt_of_num[a[l]] == 1:
                    full -= 1
                cnt_of_num[a[l]] -= 1
            sum -= a[l]
            l += 1
print(min_sum)
```

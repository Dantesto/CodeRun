# Условие

Для данных выборок $X, A, B$ найти такие $p_1,p_2,p_3$, что следующие метрики будут минимальны
$$
MSE = \frac{\sum_{i=1}^N w_i(t_i - p_1)^2}{\sum_{i=1}^N w_i}
$$
$$
MSLE = \frac{\sum_{i=1}^N w_i(\ln(1+t_i)-\ln(1+p_2))^2}{\sum_{i=1}^N w_i}
$$
$$
LogLoss = -\frac{\sum_{i=1}^N w_i\left[\frac{t_i}{C}\ln\frac{p_3}{C}+(1-\frac{t_i}{C})\ln(1-\frac{p_3}{C})\right]}{\sum_{i=1}^N w_i}
$$
где $w_i=b_i,t_i=\frac{a_i}{b_i}$, а $C$ - некоторая большая переменная, необходимая для нормировки значений $t_i,p$, которые должны $\in(0,1)$ (почему это в принципе нужно не говорится в условии, но видимо это связано с $LogLoss$ и вторым слагаемым в его формуле, где из $1$ вычитаются $\frac{t_i}{C},\frac{p_3}{C}$).

# Решение

Сразу отметим, что по всей видимости в этой задаче $X$ дан, чтобы запутать, так как в формулах не используется. В дальнейшем в коде для нас это бесполезные данные.

Для всех метрик найдём производную относительно $p$, приравняем к нулю, чтобы потом найти экстремумы, и выведем $p$.

## 1. MSE

Сразу будем подставлять в формулы $a_i,b_i$, чтобы работать с известными значениями и упрощать выражения.
$$
MSE=\frac{\sum_{i=1}^N b_i(\frac{a_i}{b_i}-p_1)^2}{\sum_{i=1}^N b_i}
$$
$$
MSE_{p_1}'=\frac{\sum_{i=1}^N b_i*2(\frac{a_i}{b_i}-p_1)*(-1)}{\sum_{i=1}^N b_i}=\frac{2\sum_{i=1}^N b_i(p_1-\frac{a_i}{b_i})}{\sum_{i=1}^N b_i}=\frac{2\left(p_1\sum_{i=1}^N b_i-\sum_{i=1}^N a_i\right)}{\sum_{i=1}^N b_i}=0
$$
$$
p_1\sum_{i=1}^N b_i-\sum_{i=1}^N a_i=0
$$
$$
p_1=\frac{\sum_{i=1}^N a_i}{\sum_{i=1}^N b_i}
$$

## 2. MSLE

Хотя только в первой функции потерь имел смысл подставлять $a_i,b_i$, здесь я все еще подставлял, не додумавшись отказаться от этого.
$$
MSLE=\frac{\sum_{i=1}^N b_i(\ln(1+\frac{a_i}{b_i})-\ln(1+p_2))^2}{\sum_{i=1}^N b_i}
$$
$$
MSLE_{p_2}'=\frac{\sum_{i=1}^N 2b_i\left(\ln(1+\frac{a_i}{b_i})-\ln(1+p_2)\right)\frac{-1}{1+p_2}}{\sum_{i=1}^N b_i}=\frac{-\frac{2}{1+p_2}\sum_{i=1}^N b_i\left(\ln(1+\frac{a_i}{b_i})-\ln(1+p_2)\right)}{\sum_{i=1}^N b_i}=-2*\frac{\sum_{i=1}^N b_i\ln(1+\frac{a_i}{b_i})-\ln(1+p_2)\sum_{i=1}^N b_i}{(1+p_2)\sum_{i=1}^N b_i}=0
$$
$$
\sum_{i=1}^N b_i\ln(1+\frac{a_i}{b_i})-\ln(1+p_2)\sum_{i=1}^N b_i=0
$$
$$
\ln(1+p_2)=\frac{\sum_{i=1}^N b_i\ln(1+\frac{a_i}{b_i})}{\sum_{i=1}^N b_i}=\overline{\ln(1+\frac{a_i}{b_i})}
$$
где $\overline{x}$ - это [среднее арифметическое взвешенное](https://ru.wikipedia.org/wiki/%D0%A1%D1%80%D0%B5%D0%B4%D0%BD%D0%B5%D0%B5_%D0%B0%D1%80%D0%B8%D1%84%D0%BC%D0%B5%D1%82%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%BE%D0%B5_%D0%B2%D0%B7%D0%B2%D0%B5%D1%88%D0%B5%D0%BD%D0%BD%D0%BE%D0%B5). Значения что это по сути не имеет, я ввел это обозначение для упрощения дальнейших записей.
$$
p_2=e^{\overline{\ln(1+\frac{a_i}{b_i})}}-1
$$

## 3. LogLoss

Здесь я уже понял, что делать замену $a_i,b_i$ не имеет смысла.
$$
LogLoss_{p_3}'=-\frac{\sum_{i=1}^N w_i\left[\frac{t_i}{C}\frac{C}{p_3}+\frac{C-t_i}{C}\frac{C}{C-P_3}*(-1)\right]}{\sum_{i=1}^N w_i}=-\frac{\sum_{i=1}^N w_i(\frac{t_i}{p_3}-\frac{C-t_i}{C-p_3})}{\sum_{i=1}^N w_i}=-\frac{\sum_{i=1}^N w_i\frac{t_iC-t_ip_3-Cp_3+t_ip_3}{p_3(C-p_3)}}{\sum_{i=1}^N w_i}=-\frac{C}{p_3(C-p_3)}\frac{\sum_{i=1}^N w_i(t_i-p_3)}{\sum_{i=1}^N w_i}=0
$$
$$
\sum_{i=1}^N w_i(t_i-p_3)=\sum_{i=1}^N w_it_i - p_3\sum_{i=1}^N w_i=0
$$
$$
p_3=\frac{\sum_{i=1}^N w_it_i}{\sum_{i=1}^N w_i}=\frac{\sum_{i=1}^N b_i\frac{a_i}{b_i}}{\sum_{i=1}^N b_i}=\frac{\sum_{i=1}^N a_i}{\sum_{i=1}^N b_i}
$$

# Код

```python
import math

n, k = [int(x) for x in input().split()]
a = []
b = []
for i in range(n):
    xab = [float(x) for x in input().split()]
    a.append(xab[-2])
    b.append(xab[-1])
ln_sum = 0
for i in range(n):
    ln_sum += b[i] * math.log1p(a[i] / b[i]) #math.log1p(x) = ln(1+x)
print(sum(a) / sum(b), math.exp(ln_sum / sum(b)) - 1, sum(a) / sum(b))
```

# Условие

Найти $a,b,c$ для

$$
f(x) = \left\{
    \begin{array}\\
        a & \mbox{if } \ x < c \\
        b & \mbox{if } \ x \geq c \\
    \end{array}
\right.
$$

минимизируя

$$ \sigma = \sqrt{\frac{\sum_{i=1}^n (f(x_i)-y_i)^2}{n}} $$

Решить при заданном наборе данных $X, Y$, т. е. проверок на других данных не будет проводиться.

# Решение

## 1. Вывод значений a, b, c

Ответ для $c$ лежит в диапазоне $[x_{min},x_{max}]$, так что будем перебирать все такие $c_i=\frac{x_{i-1}+x_i}{2}$, либо, что проще, брать $c_i= x_i$, так как в описании функции есть нестрогое неравенство. Чтобы находить $a_i$ и $b_i$, выведем их значение из функции ошибки, которую пытаемся минимизировать. Для этого приравняем производную функции потерь к нулю, где и будет её минимум. Из полученной формулы выведем искомые значения.

$L=\{(x_i,y_i):x_i<c\}$
<br>
$|L|$ is a size of a set
<br>
$a=argmin_t \sqrt{\frac{\sum_{i=1}^{|L|}(t-y_i)^2}{|L|}},$ where $y_i \in L$
<br>
$\frac{d}{dt}\sqrt{\frac{\sum_{i=1}^{|L|}(t-y_i)^2}{|L|}}=\frac{1}{\sqrt{|L|}}*\frac{1}{2\sqrt{\sum_{i=1}^{|L|}(t-y_i)^2}}*2(\sum_{i=1}^{|L|}(t-y_i))=\frac{|L|t-\sum_{i=1}^{|L|}y_i}{\sqrt{|L|\sum_{i=1}^{|L|}(t-y_i)^2}}=0$
<br>
$|L|t-\sum_{i=1}^{|L|}y_i=0$
<br>
$t=\frac{\sum_{i=1}^{|L|}y_i}{|L|}=a$
<br>
Аналогично $b$ является средним арифметическим по $y_i$ среди $R=\{(x_i,y_i):x_i\geq c\}$

## 2. Оптимизация пересчета $a_i, b_i,\sigma_i$

Для пересчета $a_i,b_i$ и $\sigma$ для каждого $c_i$ не будем пробегаться по списку чисел полностью, так как тогда решение окажется квадратичным. Будем хранить сумму левых от $c_i$ значеий $y_l$ и правых $y_r$, тогда пересчитывать $a_i,b_i$ можно сразу. А для подсчета $\sigma$ перепишем его формулу.

$$
\sigma_i=\sqrt{\frac{\sum_{j=1}^n(f(x_j)-y_j)^2}{n}}=\sqrt{\frac{\sum_{j=1}^i(a_i-y_j)^2+\sum_{j=i+1}^n(b_i-y_j)^2}{n}}=\sqrt{\frac{(i*a_i^2-2a_i\sum_{j=1}^iy_j+\sum_{j=1}^iy_j^2)+((n-i)*b_i^2-2b_i\sum_{j=i+1}^ny_j+\sum_{j=i+1}^ny_j^2)}{n}}
$$

Для пересчета $\sigma$ достаточно еще хранить сумму квадратов слева и справа, как видно по последнему слагаемому в обеих скобках в числителе.

# Код

```python
n = int(input())
data = sorted([[int(x) for x in input().split()] for i in range(n)])
sum_l, sum_l2 = 0, 0
sum_r, sum_r2 = sum(x[1] for x in data), sum(x[1]**2 for x in data)
a, b, c = 0, sum_r / n, data[0][0]
min_loss = ((n * b**2 - 2 * b * sum_r + sum_r2) / n)**0.5
i = 1
while i < n:
    while i < n and data[i][0] == data[i - 1][0]:
        sum_l += data[i - 1][1]
        sum_l2 += data[i - 1][1]**2
        sum_r -= data[i - 1][1]
        sum_r2 -= data[i - 1][1]**2
        i += 1
    if i == n:
        break
    sum_l += data[i - 1][1]
    sum_l2 += data[i - 1][1]**2
    sum_r -= data[i - 1][1]
    sum_r2 -= data[i - 1][1]**2
    ci = data[i][0]
    ai = sum_l / i
    bi = sum_r / (n - i)
    lossi = (((i * ai**2 - 2 * ai * sum_l + sum_l2) + ((n - i) * bi**2 - 2 * bi * sum_r + sum_r2)) / n)**0.5
    if lossi < min_loss:
        a, b, c, min_loss = ai, bi, ci, lossi
    i += 1
print(a, b, c)
```
